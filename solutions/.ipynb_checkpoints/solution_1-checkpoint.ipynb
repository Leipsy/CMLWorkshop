```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
```

**Exercise: Load power T-stats from an example subject ("R1001P_powers.npy") and create a time-frequency plot of electrode 70**

```python
# Load the data file R1001P_tstats.npy from the npy_files directory:
# The os.path.join command concatenates the sample_files subdirectory with the filename
# in an OS-independent manner.
# Resulting data dimensions in t are [frequencies, electrodes, time bins].
import os
t = np.load(os.path.join('../sample_files', 'R1001P_tstats.npy'))
```

```python
plotdata = t[:,69,:]
plt.matshow(plotdata, fignum=0, aspect='auto', cmap='RdBu_r', vmin=-3.5, vmax=3.5)
plt.colorbar()
```

**Exercise: Create a time-frequency plot where only power values exceeding a T-stat of 2 are shown**

```python
filtered_data = [[x if x > 2 else 0 for x in y] for y in plotdata]
plt.matshow(filtered_data, fignum=0, aspect='auto', cmap='RdBu_r', vmin=-3.5, vmax=3.5)
plt.colorbar()

# Alternate
threshold_dat = t[:, 69, :]>2
plt.matshow(threshold_dat, fignum=0, aspect='auto', cmap='RdBu_r', vmin=0, vmax=1)

# Alternate
electrode = t[:, 69, :]
plt.matshow(np.where(electrode > 2, electrode, 0), fignum=0, aspect='auto', cmap='RdBu_r', vmin=-3.5, vmax=3.5)
plt.colorbar()
```

# Assignment 1

```python
# 1
ch = pd.read_csv(os.path.join('../sample_files', 'R1292E_FR1_0_channels.csv'))
mask = [isinstance(x, str) and 'CA1' in x for x in ch['stein.region']]
ca1ch = ch[mask]
print(len(ca1ch), 'CA1 channels')
print()
```

```python
# 2
print('Region      ', '\t', 'mni.x')
print('-'*40)
for k in ca1ch.index:
  print(ca1ch.loc[k, 'stein.region'], '\t', ca1ch.loc[k, 'mni.x'])
print()
```

```python
# 3
left = ca1ch[ca1ch['mni.x']<0]
right = ca1ch[ca1ch['mni.x']>0]

def AvgCoords(chs):
  def MeanStd(chs, axstr):
    arr = np.array(chs['mni.'+axstr])
    print('mu_'+axstr+': ', np.mean(arr), ', sigma_'+axstr+': ', np.std(arr, ddof=1), sep='')

  MeanStd(chs, 'x')
  MeanStd(chs, 'y')
  MeanStd(chs, 'z')

print('Left:')
AvgCoords(left)
print('Right:')
AvgCoords(right)
print()
```

```python
# 4
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(left['mni.x'], left['mni.y'], left['mni.z'])
ax.scatter(right['mni.x'], right['mni.y'], right['mni.z'])
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')
plt.show()
```
