```python
# 1
def SerPosProb(dfr):
  evs = load.Load(dfr, 'events')
  evs_filt = evs[evs['list']>0]
  evs_filt = evs_filt[evs_filt['serialpos']>0]
  recvals = np.array(evs_filt.groupby('serialpos')['recalled'].sum())
  totals = evs_filt.groupby('serialpos').size()
  return recvals / totals

df_subexp = df[(df['subject']=='R1111M') & (df['experiment']=='FR1')]
dfr = df_subexp.iloc[0]
recprob = SerPosProb(dfr)

plt.figure()
plt.plot(recprob)
plt.xlabel('Serial position')
plt.ylabel('Recall probability')
plt.title('R1111M, FR1, session 0')
plt.show()
```

```python
# 2

sub = 'R1292E'
exp = 'FR1'
df_subexp = df[(df['subject']==sub) & (df['experiment']==exp)]
dfr = df_subexp.iloc[0]

evs = load.Load(dfr, 'events')
evs_rec = evs[evs['type']=='REC_WORD']
rec_order_gb = evs_rec.groupby('list')['serialpos']
rec_order = np.array(list(rec_order_gb.apply(list)))

# Don't count lags to repeats / intrusions
rec_order = [[e if e>0 else np.nan for e in r] for r in rec_order]
lags = [np.diff(a) for a in rec_order]
lags = [np.array([L for L in lagarr if not np.isnan(L)], dtype=int) for lagarr in lags]
rec_filtered = [[r1 for r1,r2 in zip(rarr,rarr[1:]) if not np.isnan(r2)] for rarr in rec_order]

def PossibTransit(rec_filtered, maxval):
  remaining = np.ones(maxval, dtype=bool)
  possib = np.zeros(2*maxval-1, dtype=int)
  for r in rec_filtered:
    if np.isnan(r):
      continue
    remaining[r-1] = False
    possib[np.argwhere(remaining).ravel()-(r-1)+maxval-1] += 1
  return possib

list_sz = len({e for e in evs['serialpos'] if e>0})
poss_arr = [PossibTransit(r, list_sz) for r in rec_filtered]
lag_arr = [np.bincount(L+list_sz-1, minlength=2*list_sz-1) for L in lags]
poss_arr = np.sum(poss_arr, axis=0)
lag_arr = np.sum(lag_arr, axis=0)
cond_prob = [Le/Lp if Lp>0 else 0 for Le,Lp in zip(lag_arr, poss_arr)]

plt.figure()
plt.plot(np.arange(-list_sz+1,0), cond_prob[:list_sz-1])
plt.plot(np.arange(1,list_sz), cond_prob[list_sz:])
plt.xlabel('Lag')
plt.ylabel('Conditional Response Probability')
plt.title(sub+', '+exp+', session 0')
plt.show()
```
